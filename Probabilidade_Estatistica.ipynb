```python
# ==============================================
#  RELATÓRIO DE PROBABILIDADE E ESTATÍSTICA - PNAD
# ==============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
from scipy.stats import chi2_contingency

# ----------------------------------------------
# 1) Carregar os dados
# ----------------------------------------------
df = pd.read_csv("arquivo.csv.xls")

# ----------------------------------------------
# 2) Dicionário de variáveis categóricas
# ----------------------------------------------
data_dict = {
    'UF': {
        11: 'Rondônia', 12: 'Acre', 13: 'Amazonas', 14: 'Roraima', 15: 'Pará', 16: 'Amapá', 17: 'Tocantins',
        21: 'Maranhão', 22: 'Piauí', 23: 'Ceará', 24: 'Rio Grande do Norte', 25: 'Paraíba', 26: 'Pernambuco',
        27: 'Alagoas', 28: 'Sergipe', 29: 'Bahia', 31: 'Minas Gerais', 32: 'Espírito Santo', 33: 'Rio de Janeiro',
        35: 'São Paulo', 41: 'Paraná', 42: 'Santa Catarina', 43: 'Rio Grande do Sul', 50: 'Mato Grosso do Sul',
        51: 'Mato Grosso', 52: 'Goiás', 53: 'Distrito Federal'
    },
    'Sexo': {0: 'Masculino', 1: 'Feminino'},
    'Cor': {0: 'Indígena', 2: 'Branca', 4: 'Preta', 6: 'Amarela', 8: 'Parda', 9: 'Sem declaração'},
    'Anos de Estudo': {
        1: 'Sem instrução e menos de 1 ano', 2: '1 ano', 3: '2 anos', 4: '3 anos', 5: '4 anos', 6: '5 anos',
        7: '6 anos', 8: '7 anos', 9: '8 anos', 10: '9 anos', 11: '10 anos', 12: '11 anos', 13: '12 anos',
        14: '13 anos', 15: '14 anos', 16: '15 anos ou mais', 17: 'Não determinados'
    }
}

for col, mapping in data_dict.items():
    if col in df.columns:
        df[col] = df[col].map(mapping)

# =======================================================
print("\n\033[1m========== 1) CLASSIFICATIPOS DE VARIÁVEIS ==========\033[0m\n")
# =======================================================

print("""
Nesta etapa, classifico as variáveis de acordo com suas características.
As variáveis qualitativas nominais são aquelas sem ordem natural (como UF, Sexo e Cor).
A variável qualitativa ordinal (Anos de Estudo) possui uma hierarquia —
mais anos de estudo representam maior nível de instrução.
Já as variáveis quantitativas são aquelas que podem ser medidas
numericamente: Idade (discreta), Renda e Altura (contínuas).
""")

print("Variáveis Qualitativas Nominais: UF, Sexo, Cor")
print("Variáveis Qualitativas Ordinais: Anos de Estudo")
print("Variáveis Quantitativas Discretas: Idade")
print("Variáveis Quantitativas Contínuas: Renda, Altura")
print("-" * 80)

# =======================================================
print("\n\033[1m========== 2) DISTRIBUIÇÃO DE FREQUÊNCIA ==========\033[0m\n")
# =======================================================
print("""
Aqui analiso como os valores se distribuem dentro de cada categoria.
Essa etapa ajuda a identificar qual grupo tem maior representatividade na amostra.
Por exemplo, se há predominância de um sexo, cor ou estado específico.
A distribuição percentual facilita a comparação entre categorias.
""")

for col in ['UF', 'Sexo', 'Cor', 'Anos de Estudo']:
    print(f"\n Distribuição de {col}:")
    freq_table = df[col].value_counts(normalize=True).mul(100).round(2).astype(str) + '%'
    print(freq_table)
print("-" * 80)

# =======================================================
print("\n\033[1m========== 3) GRÁFICOS ==========\033[0m\n")
# =======================================================
print("""
Os gráficos de barras mostram a proporção das variáveis qualitativas, como Sexo, Cor e UF.
Os histogramas exibem a distribuição das quantitativas, como Idade, Renda e Altura.
É possível observar padrões visuais — como faixas etárias mais comuns, concentração de rendas ou variação da altura média.
""")

plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
sns.countplot(data=df, x='Sexo', palette='viridis')
plt.title('Distribuição de Sexo')
plt.subplot(1, 3, 2)
sns.countplot(data=df, x='Cor', palette='viridis')
plt.title('Distribuição de Cor')
plt.subplot(1, 3, 3)
sns.countplot(data=df, x='UF', palette='viridis')
plt.title('Distribuição por UF')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
print("-" * 80)

# =======================================================
print("\n\033[1m========== 4) MEDIDAS DE POSIÇÃO ==========\033[0m\n")
# =======================================================
print("""
Analiso os valores centrais de cada variável quantitativa.

- Média representa o valor médio geral.

- Mediana indica o ponto central (metade dos valores estão acima e metade abaixo).

- Moda mostra o valor mais frequente.

Comparar essas três medidas ajuda a entender se os dados estão simétricos ou assimétricos
(por exemplo, renda costuma ser assimétrica, com poucos valores muito altos).
""")

for col in ['Idade', 'Renda', 'Altura']:
    media = df[col].mean()
    mediana = df[col].median()
    moda = df[col].mode()[0]
    print(f"\n {col.upper()}")
    print(f"Média: {media:.2f}")
    print(f"Mediana: {mediana:.2f}")
    print(f"Moda: {moda:.2f}")
    if media > mediana:
        print("→ Assimetria à direita (poucos valores muito altos elevam a média).")
    elif media < mediana:
        print("→ Assimetria à esquerda (valores baixos puxam a média).")
    else:
        print("→ Distribuição simétrica.")
print("-" * 80)

# =======================================================
print("\n\033[1m========== 5) MEDIDAS DE DISPERSÃO ==========\033[0m\n")
# =======================================================
print("""
As medidas de dispersão indicam o quanto os valores variam.

- Variância e Desvio Padrão mostram o grau de dispersão.

- Amplitude é a diferença entre o valor máximo e o mínimo.

- Coeficiente de Variação (CV) indica a variação relativa em relação à média
(CV > 30% sugere alta dispersão).

Isso ajuda a perceber, por exemplo, se há grande desigualdade na renda ou na altura dos indivíduos.
""")

for col in ['Idade', 'Renda', 'Altura']:
    var = df[col].var()
    std = df[col].std()
    amplitude = df[col].max() - df[col].min()
    cv = (std / df[col].mean()) * 100
    print(f"\n {col.upper()}")
    print(f"Variância: {var:.2f}")
    print(f"Desvio Padrão: {std:.2f}")
    print(f"Amplitude: {amplitude:.2f}")
    print(f"Coeficiente de Variação: {cv:.2f}%")
    if cv > 30:
        print("→ Alta dispersão nos dados.")
    else:
        print("→ Baixa dispersão, dados homogêneos.")
print("-" * 80)

# =======================================================
print("\n\033[1m========== 6) QUANTIS ==========\033[0m\n")
# =======================================================
print("""
Os quantis (25%, 50%, 75%) dividem os dados em quatro partes (quartis).

- O 1º quartil (25%) indica o limite inferior.

- O 2º quartil (50%) é a mediana.

- O 3º quartil (75%) mostra o limite superior.

Eles são úteis para compreender a dispersão e comparar intervalos de dados.
""")
for col in ['Idade', 'Renda', 'Altura']:
    print(f"\n Quantis de {col}:")
    print(df[col].quantile([0.25, 0.5, 0.75]))
print("-" * 80)

# =======================================================
print("\n\033[1m========== 7) BOX PLOTS ==========\033[0m\n")
# =======================================================
print("""
O boxplot representa graficamente os quartis e possíveis valores discrepantes (outliers).
Ele mostra onde os dados se concentram e onde ocorrem variações extremas,
sendo especialmente útil para comparar a distribuição da renda ou altura entre grupos.
""")
plt.figure(figsize=(15, 5))
for i, col in enumerate(['Idade', 'Renda', 'Altura']):
    plt.subplot(1, 3, i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Box Plot de {col}')
plt.tight_layout()
plt.show()
print("-" * 80)

# =======================================================
print("\n\033[1m========== 8) ASSOCIAÇÃO ENTRE VARIÁVEIS QUALITATIVAS ==========\033[0m\n")
# =======================================================
print("""
O teste de Qui-quadrado avalia se existe relação entre duas variáveis categóricas — no caso, Sexo e Cor.

- Um valor-p menor que 0,05 indica associação estatisticamente significativa.

- Caso contrário, não há evidência de relação entre as categorias.

O gráfico de barras agrupadas ajuda a visualizar se os grupos diferem entre si.
""")

crosstab = pd.crosstab(df['Sexo'], df['Cor'])
chi2, p, dof, expected = chi2_contingency(crosstab)
print("Tabela de Contingência (Sexo vs Cor):")
print(crosstab)
print(f"\nEstatística Qui-quadrado: {chi2:.2f}")
print(f"Valor p: {p:.3f}")
if p < 0.05:
    print("→ Há associação significativa entre Sexo e Cor.")
else:
    print("→ Não há associação significativa entre Sexo e Cor.")
print("-" * 80)

# =======================================================
print("\n\033[1m========== 9) ASSOCIAÇÃO ENTRE VARIÁVEIS QUANTITATIVAS ==========\033[0m\n")
# =======================================================
print("""
A correlação de Pearson mede o grau de relação linear entre duas variáveis numéricas (como Idade e Renda).

- Se o coeficiente for próximo de 1 ou -1, a relação é forte.

- Se for próximo de 0, é fraca ou inexistente.

O gráfico de dispersão mostra visualmente se a renda tende a crescer com a idade ou não.
""")

corr = df[['Idade', 'Renda', 'Altura']].corr()
print("Matriz de Correlação:")
print(corr)
print("\nAnálise da correlação entre Idade e Renda:")
corr_idade_renda = df['Idade'].corr(df['Renda'])
print(f"Coeficiente de Correlação: {corr_idade_renda:.2f}")
if abs(corr_idade_renda) > 0.5:
    print("→ Correlação forte.")
elif abs(corr_idade_renda) > 0.2:
    print("→ Correlação moderada.")
else:
    print("→ Correlação fraca ou inexistente.")
print("-" * 80)

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Idade', y='Renda', alpha=0.5)
plt.title('Associação entre Idade e Renda')
plt.show()

# =======================================================
print("\n\033[1m CONCLUSÃO GERAL\033[0m\n")
# =======================================================
print("""
Com base nos resultados, é possível observar as principais tendências da amostra:

- Distribuição das categorias sociais e educacionais;

- Padrões de dispersão em renda e altura;

- Possíveis relações entre variáveis, como idade e rendimento.

Essa análise fornece um panorama estatístico completo e fundamenta interpretações futuras com base em evidências quantitativas.
""")

print("\n\033[1m Autor Jade Pereira da Paz. \033[0m ")
```
